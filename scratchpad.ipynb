```python
import networkx as nx
import matplotlib as plt
import random
import math

%matplotlib inline

plt.rcParams['figure.figsize'] = [20, 10]

def people_names():
    import faker
    import faker.providers
    import faker.providers.person
    from faker.providers.person.en import Provider
    
    return Provider.first_names
```

```python
def friends_approves():
    G = nx.DiGraph()
    
    requestor = random.choice(people_names())
    identity_approvers = random.choices(people_names(), k=3)
    crowd = random.choices(people_names(), k=20)
    
    G.add_node(requestor)
    G.add_nodes_from(identity_approvers)
    G.add_nodes_from(crowd)
    
    for account in identity_approvers:
        G.add_edge(account, requestor)
    
    return G
```

```python
G = friends_approves()
Gprim = max(nx.weakly_connected_component_subgraphs(G), key=len)
print(Gprim)
nx.draw_networkx(Gprim)
print(nx.pagerank(G, alpha=0.9))
```

```python
def friends_approves2():
    G = nx.DiGraph()
    
    requestor = random.choice(people_names())
    identity_approvers = random.choices(people_names(), k=3)
    friends_of_identity_approvers = random.choices(people_names(), k=11)
    crowd = random.choices(people_names(), k=20)
    
    G.add_node(requestor)
    G.add_nodes_from(identity_approvers)
    G.add_nodes_from(friends_of_identity_approvers)
    G.add_nodes_from(crowd)
    
    for account in identity_approvers:
        G.add_edge(account, requestor)
    
    for account in identity_approvers:
        approvers = random.choices(friends_of_identity_approvers, k=3)
        for approver in approvers:
            G.add_edge(approver, account)
    
    return G
```

```python
G = friends_approves2()
Gprim = max(nx.weakly_connected_component_subgraphs(G), key=len)
print(Gprim)
nx.draw_networkx(Gprim)
print(nx.pagerank(G, alpha=0.9))
```

```python
def cycle(k=4):
    G = nx.DiGraph()
    
    requestor = random.choice(people_names())
    identity_approvers = random.choices(people_names(), k=k)
    crowd = random.choices(people_names(), k=20)
    
    G.add_node(requestor)
    G.add_nodes_from(identity_approvers)
    G.add_nodes_from(crowd)
    
    G.add_edge(identity_approvers[-1], requestor)
    G.add_cycle(identity_approvers)
    
    return G
```

```python
G = cycle()
Gprim = max(nx.weakly_connected_component_subgraphs(G), key=len)
print(Gprim)
nx.draw_networkx(Gprim)
print(nx.pagerank(G, alpha=0.85, max_iter=22))
```

```python
def bots_full(n=5):
    G = nx.DiGraph()
    
    bots = random.choices(people_names(), k=n)
    crowd = random.choices(people_names(), k=20)
    
    G.add_nodes_from(bots)
    G.add_nodes_from(crowd)
    
    for bot in bots:
        for another_bot in bots:
            G.add_edge(bot, another_bot)
    
    return G
```

```python
def trust_rank(g: nx.DiGraph, alpha=0.85) -> dict:
    def estimate_trust(node, level=0, visited_nodes=None):
        visited_nodes = visited_nodes or set()
        predecessors = set(g.predecessors(node)).difference(visited_nodes)
        
        if level >= 3:
            return 0
        elif level == 2:
            return alpha * math.sqrt(1 + len(predecessors)) - 1
        else:
            trusts = [estimate_trust(predecessor, level=level+1, visited_nodes={node, predecessor}.union(visited_nodes))
                     for predecessor in predecessors if predecessor not in visited_nodes]
            avg_trust = (sum(trusts) / len(trusts)) if len(trusts) > 0 else 0
            return alpha * (avg_trust + math.sqrt(1 + sum((trust - avg_trust) ** 2 for trust in trusts if trust > avg_trust)))
    
    return {node: estimate_trust(node) - alpha for node in g.nodes}
```

```python
G = bots_full(40)
Gprim = max(nx.weakly_connected_component_subgraphs(G), key=len)
print(Gprim)
nx.draw_networkx(Gprim)
print(trust_rank(G))
```

```python
def bots_circles(k=2, n=5):
    G = nx.DiGraph()
    
    bots = random.choices(people_names(), k=n)
    crowd = random.choices(people_names(), k=20)
    
    G.add_nodes_from(bots)
    G.add_nodes_from(crowd)
    
    for _ in range(k):
        path = list(bots)
        random.shuffle(bots)
        
        for i, _ in enumerate(path):
            G.add_edge(path[i-1], path[i])
    
    return G
```
